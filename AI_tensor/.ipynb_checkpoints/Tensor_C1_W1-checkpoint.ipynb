```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
print(tf.__version__)
```

## Define and Compile the Neural Network

##### Create simplest possible NN: 
##### 1. It has 1 layer 1 neuron and the input shape of 1. We will build this model using Keras` Sequential class which allows to #####define the newtwork as a sequence of layers.We use single Dense layer to build this simple network

```python
## Build a simple sequential model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

##### Now we need to compile the NN. We need to do 2 things: specify 2 functions - a loss function and an optimizer 
##### when the computer try to learn the relationship between x and y, at first it learns by guessing.
The loss function measures the guessed answers against the known correct answers and quantify how bad or well it did.
it then uses optimizer function to make another guess. Based on how the loss function went, it will try to minimize the loss.
It will repeat this for the number of epoches.Let`s use mean squared error to measure the loss, and stochastic gradient descent for the optimizer

```python
## compile the model
model.compile(optimizer= 'sgd', loss ='mean_squared_error')
```

## Providing the data


```python
### Declare the model inputs and outputs for training
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0,-1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Training the Neural Network

### the process of training the NN, where it 'learns ' the relationship betweeb x and y is in the model.fit().
This is where it will go through the loop of making a guess, measuring loss  and optimizing for another guess. 

```python
# train the model
model.fit(xs, ys, epochs=500)
```

## Make predictions

```python
print (model.predict([10.0, 15.0, -5.0]))
```
